```python
import pandas as pd
import numpy as np


Y =  [78.5, 74.3, 104.3, 87.6, 95.9, 109.2, 102.7, 72.5, 93.1, 115.9, 83.8, 113.3, 109.4]
X1 = [7, 1, 11, 11, 7, 11, 3, 1, 2, 21, 1, 11, 10]
X2 = [26, 29, 56, 31, 52, 55, 71, 31, 54, 47, 40, 66, 68]
X3 = [6, 15, 8, 8, 6, 9, 17, 22, 18, 4, 23, 9, 8]
X4 = [60, 52, 20, 47, 33, 22, 6, 44, 22, 26, 34, 12, 12]

df = pd.DataFrame({
    "Y": Y,
    "X1": X1,
    "X2": X2,
    "X3": X3,
    "X4": X4
})
df
```

```python
df2 = df.copy()
df2['Y']=1
df2
```

```python
X = df2.values
print(X)
```

```python
# Find the transpose of the x matrix
Xt = X.T

# Display the transposed matrix
print("Transpuesta de x:")
print(Xt)
```

```python
# Calculate the matrix product of x_transpose and x
xt_x = Xt @ X

# Calculate the determinant of xt_x
determinant = np.linalg.det(xt_x)

# Display the determinant
print("Determinante de (x_transpose * x):")
display(determinant)
print(xt_x)
```

```python
# Calculate the inverse of xt_x
xt_x_inv = np.linalg.inv(xt_x)

# Display the inverse matrix
print("Inversa de (x_transpose * x):")
print(xt_x_inv)
```

```python
y= df['Y'].values
betas = xt_x_inv @ Xt @ y
betas
```

```python
########## CALCULO DEL ERROR #######################
df['Y_pred'] = betas[0] + betas[1]*df['X1'] + betas[2]*df['X2'] + betas[3]*df['X3'] + betas[4]*df['X4']
df
```

```python
df['Error'] = df['Y'] - df['Y_pred']
df
```

```python
df['Error^2'] = df['Error']**2
df
```

```python
ErrorTotal = df['Error^2'].sum()
print(ErrorTotal)
```
