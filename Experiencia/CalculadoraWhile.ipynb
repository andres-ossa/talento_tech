```python
# Importando la librería de matemáticas
import math

# --- Definición de funciones ---
def mostrar_titulo():
    print('Calculadora Sencilla (+,-,*,/)')

def mostrar_menu():
    print()
    print("MENU:")
    print("1: Sumar (+)")
    print("2: Restar (-)")
    print("3: Multiplicar (*)")
    print("4: Dividir (/)")
    print("5: Logaritmo (log10)")
    print("6: Coseno (cos)")
    print("7: Potenciacion")
    print("8: Radicacion")
    print("9: Seno (sin)")
    print("10: Tangente (tan)")
    print("11: Salir")
    print()

def ingresar_datos():
    numero_uno = float(input('Ingrese el primer número de la operación: '))
    numero_dos = float(input('Ingrese el segundo número de la operación: '))
    return numero_uno, numero_dos

# --- Operaciones ---
def calcular_suma(a,b): return a+b
def calcular_resta(a,b): return a-b
def calcular_multiplicacion(a,b): return a*b
def calcular_division(a,b): return a/b
def calcular_logaritmo_base10(x): return math.log10(x)
def calcular_coseno(x): return math.cos(x)
def calcular_potenciacion(base,exp): return base**exp
def calcular_radicacion(num,indice): return "Error: índice 0" if indice==0 else num**(1/indice)
def calcular_seno(x): return math.sin(x)
def calcular_tangente(x): return math.tan(x)

# --- Procesamiento ---
def procesar_y_mostrar_operacion(opcion, numero_uno, numero_dos):
    if opcion == '1':
        print("Resultado =", calcular_suma(numero_uno, numero_dos))
    elif opcion == '2':
        print("Resultado =", calcular_resta(numero_uno, numero_dos))
    elif opcion == '3':
        print("Resultado =", calcular_multiplicacion(numero_uno, numero_dos))
    elif opcion == '4':
        if numero_dos != 0:
            print("Resultado =", calcular_division(numero_uno, numero_dos))
        else:
            print("¡No se puede dividir entre cero!")
    elif opcion == '5':
        print("Log10(",numero_uno,")=",calcular_logaritmo_base10(numero_uno))
        print("Log10(",numero_dos,")=",calcular_logaritmo_base10(numero_dos))
    elif opcion == '6':
        print("Cos(",numero_uno,")=",calcular_coseno(numero_uno))
        print("Cos(",numero_dos,")=",calcular_coseno(numero_dos))
    elif opcion == '7':
        print("Resultado =", calcular_potenciacion(numero_uno, numero_dos))
    elif opcion == '8':
        print("Resultado =", calcular_radicacion(numero_uno, numero_dos))
    elif opcion == '9':
        print("Sin(",numero_uno,")=",calcular_seno(numero_uno))
        print("Sin(",numero_dos,")=",calcular_seno(numero_dos))
    elif opcion == '10':
        print("Tan(",numero_uno,")=",calcular_tangente(numero_uno))
        print("Tan(",numero_dos,")=",calcular_tangente(numero_dos))
    else:
        print("Opción no válida")

# --- Ejecución ---
mostrar_titulo()
salir = False

while not salir:
    mostrar_menu()
    opcion = input("Ingrese la opción deseada del menú: ")

    if opcion == '11':  # salir
        print("Gracias por usar la calculadora. ¡Hasta luego!")
        salir = True
    else:
        numero_uno, numero_dos = ingresar_datos()
        procesar_y_mostrar_operacion(opcion, numero_uno, numero_dos)
```
