```python
# Calculate IMC (BMI) and add it as a new column
df['imc'] = df['peso'] / (df['altura'] ** 2)

df
```

```python
def classify_imc(imc):
    if imc < 18.5:
        return 'Bajo peso'
    elif 18.5 <= imc < 24.9:
        return 'Peso normal'
    elif 25.0 <= imc < 29.9:
        return 'Sobrepeso'
    else:
        return 'Obesidad'

df['clasificacion_imc'] = df['imc'].apply(classify_imc)

# Display the updated DataFrame with the new 'clasificacion_imc' column
df
```
